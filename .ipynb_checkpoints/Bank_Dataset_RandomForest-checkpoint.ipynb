#### Here if one hot encoding is used then 45 new columns would be created.

#### Hence lets use mean encoding to change the categorical variables into numerical variables. 

```python
df1=df.copy()
```

```python
### Defining Mean Encoding.
def Mean_encoding(df,datacolumn,target):
    mean_ordinal=df.groupby(datacolumn)[target].mean().to_dict()
    df['mean_encode '+datacolumn]=df[datacolumn].map(mean_ordinal)
    df=df.drop(datacolumn,axis=1)    
    return df
```

```python
### Using Mean Encoding technique to replace the values in object_dtype_columns into float values.

for i in object_dtype_columns:
    df1=Mean_encoding(df1,i,'y')
```

```python
df1.y.value_counts()
```

```python
## Checking whether the dataset is an imbalanced dataset or not.
df1.y.value_counts()/df1.shape[0]*100
```

#### Here we can clearly see that the dataset is highly imbalanced.
#### Hence lets adopt the over sampling technique to convert this imbalanced dataset into a balanced one. 

### Lets first split the dataset into subcribed to term deposit and not subcribed to term deposit.

```python
subscribed=df1[df1.y==1]
not_subscribed=df1[df1.y==0]
```

```python
subscribed.shape,not_subscribed.shape
```

```python
X=df1.drop(['y'],axis=1)
y=df1['y']
```

```python
### Using imblearn library to convert the imbalanced dataset into balanced one. 

from imblearn.combine import SMOTETomek
smk=SMOTETomek(random_state=42)
X_res,y_res=smk.fit_sample(X,y)
```

```python
X_res.shape,y_res.shape
```

```python
from collections import Counter
print('Original dataset shape {}'.format(Counter(y)))
print('Reshaped dataset shape {}'.format(Counter(y_res)))
```

#### Here we can clearly see that the dataset that was fed has been balanced.

```python
df2=pd.concat([X_res,y_res],axis=1)
```
